### **Organizando os dados**

<u>Renomeando variáveis (colunas)</u>

```python
covid_sp.head()
```

```python
# Renomeando as colunas apenas para visualização

covid_sp.rename(columns={'datahora': 'hora'})
```

```python
# Renoameando as colunas sobrepondo o dataframe

covid_sp = covid_sp.rename(columns={'nome_munic': 'municipio'})
```

```python
covid_sp.head()
```

```python
# Renomeando a coluna sem sobrepor o dataframe

covid_sp.rename(columns={'datahora': 'hora'}, inplace=True)
```

```python
covid_sp.rename(columns={'map_leg': 'rotulo_mapa', 'map_leg_s': 'codigo_mapa'}, inplace=True)
```

```python
covid_sp.head()
```

<u>Excluindo variáveis (colunas)</u>

```python
# Excluir por nome da coluna sobrepondo o dataframe

covid_sp_alterado = covid_sp.drop(columns=['cod_ra'])
```

```python
# Excluir mais de uma coluna por nome sem sobrepor o dataframe

covid_sp_alterado.drop(columns=['rotulo_mapa', 'codigo_mapa', 'cod_drs'], inplace=True)
```

```python
# Excluir por número do indice da coluna sobrepondo o dataframe
# axis=1 = coluna
# axis=0 = linha

covid_sp_alterado = covid_sp_alterado.drop(covid_sp_alterado.columns[[1]], axis=1)
```

```python
# Excluindo colunas pelo index sem sobrepor o dataframe

covid_sp_alterado.drop(covid_sp_alterado.columns[[13, 14, 18, 19]], axis=1, inplace=True)
```

<u>Criando e alterando os valores das colunas (variáveis)</u>

```python
covid_sp_alterado['area'] = covid_sp_alterado['area'] / 100

# ou

# covid_sp_alterado.area = covid_sp_alterado.area / 100
```

```python
covid_sp_alterado.head()
```

```python
# Criação de uma nova coluna com a densidade demográfica (hab/km2)

# covid_sp_alterado['densidade'] = covid_sp_alterado['pop'] / covid_sp_alterado['area']

#ou

# Criar a coluna em uma posição(index) especifica(o) no dataframe
covid_sp_alterado.insert(loc=17, column= 'densidade', value=covid_sp_alterado['pop'] / covid_sp_alterado['area'])
```

```python
# Criando a coluna indice

lista = list(range(1, 374035))
```

```python
# Criando um dataframe com a lista

df = pd.DataFrame(lista, columns=['indice'])
df
```

```python
# Incluindo o dataframe que é um indice, no dataframe covid_sp_alterado

covid_sp_alterado = pd.concat([covid_sp_alterado, df], axis=1)
```

```python
covid_sp_alterado.head()
```

```python
# Passando a coluna indice, que é a ultima coluna do dataframe, para ser a primeira coluna

covid_sp_alterado = covid_sp_alterado.reindex(columns=['indice'] + list(covid_sp_alterado.columns[:-1]))
```

```python
covid_sp_alterado.head()
```

<p>Otimizando o código: não precisa nem criar o DataFrame "df", insere a lista que será uma nova coluna direto no DataFrame "covid_sp_alterado", para isso podemos utilizar o código abaixo para inserir essa nova coluna no DataFrame covid_sp_alterado, sem a necessidade usar o concat e reindex.</p>

<p>Dessa forma, não será necessário posteriormente ter que alterar a posição da coluna indice.</p>

<p>Saliento que a lista tem que ter a mesma quantidade de linhas do DataFrame "covid_sp_alterado".</p>

**Insere a nova coluna diretamente no início**

covid_sp_alterado.insert(0, 'indice', lista)

### **Contagem de registros das variaveis (colunas)**

```python
# Contando a quantidade de casos por semana(epidemiologica)

covid_sp_alterado['semana_epidem'].value_counts()
```

```python
# Contado a quantidade de casos por semana (epidemiologica) e ordenando por semana

covid_sp_alterado['semana_epidem'].value_counts().sort_index()
```

```python
from collections import Counter
```

```python
Counter(covid_sp_alterado['semana_epidem'])
```

```python
# Usando a função query() para selecionar todas as linhas onde na coluna obitos_novos o valor seja maior que 50
# Depois passando a coluna municipio, para selecionar apenas os municipios que tem o resultado acima (obitos_novos > 50)
# Em seguida conta quantas vezes cada município aparece na lista filtrada.

covid_sp_alterado.query('obitos_novos > 50')['municipio'].value_counts()
```
